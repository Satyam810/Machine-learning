| Type                                     | Description                                 | Example Use                           |
| ---------------------------------------- | ------------------------------------------- | ------------------------------------- |
| **Feedforward Neural Network (FNN)**     | Simplest type; data flows in one direction. | Basic prediction tasks                |
| **Convolutional Neural Network (CNN)**   | Specializes in images and spatial data.     | Image recognition, medical imaging    |
| **Recurrent Neural Network (RNN)**       | Processes sequential data (time-dependent). | Speech recognition, text analysis     |
| **Transformer**                          | Uses attention mechanisms for sequences.    | ChatGPT, translation, text generation |
| **Autoencoder**                          | Learns efficient representations of data.   | Noise reduction, feature compression  |
| **GAN (Generative Adversarial Network)** | Generates new, realistic data.              | Deepfake creation, art generation     |

**Applications of Deep Learning**
*   Image & speech recognition
*   Natural language processing (ChatGPT!)
*   Self-driving cars
*   Healthcare (disease detection)
